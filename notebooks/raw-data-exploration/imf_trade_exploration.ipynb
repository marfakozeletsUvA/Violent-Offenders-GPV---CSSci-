```python
import pandas as pd 

df = pd.read_csv('../../data/raw/raw_imf_trade.csv')
```

```python
df.head()
```

```python
print(df.shape)
print(df.columns)
print(df.isna().sum().to_string())
```

```python
cdf = df.copy()

cdf = cdf.drop(columns=['FREQUENCY', 'SCALE'])

cdf.rename(columns={
    'COUNTRY': 'country_a',
    'INDICATOR': 'import_export',
    'COUNTERPART_COUNTRY': 'country_b',
    'TIME_PERIOD': 'year',
    'OBS_VALUE': 'value_mill_dollar'
}, inplace=True)

cdf.head()
```

```python

cdf['import_export'] = cdf['import_export'].replace('Imports of goods, Free on board (FOB), US dollar', 'import')
cdf['import_export'] = cdf['import_export'].replace('Exports of goods, Free on board (FOB), US dollar', 'export')

print("Import count:", sum(cdf['import_export'] == 'import'))
print("Export count:", sum(cdf['import_export'] == 'export'))
```

```python
unique = cdf['year'].unique()
print(unique)
```

```python
# Remove the last character from all values in a column
cdf['year'] = cdf['year'].astype(str).str[:-1]

unique = cdf['year'].unique()
print(unique)
```

```python
# Select rows where COUNTRY is 'USA'
nan_rows = cdf[cdf['year'] == 'n']
print(nan_rows)
```

```python
cdf = cdf.dropna(subset=['value_mill_dollar'])
print(nan_rows)
```
