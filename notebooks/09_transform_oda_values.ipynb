```python
import pandas as pd

df = pd.read_csv('../data/merged/panel_with_controls_1992_2024.csv')

# Check before
neg_mask = df['bilateral_oda'] < 0
print(f'Negative rows: {neg_mask.sum():,} ({neg_mask.mean()*100:.1f}%)')

# Floor at 0
df['bilateral_oda'] = df['bilateral_oda'].clip(lower=0)

assert df['bilateral_oda'].min() >= 0
print('Done. Min bilateral_oda:', df['bilateral_oda'].min())

df.to_csv('../data/merged/panel_with_controls_1992_2024.csv', index=False)
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/merged/panel_with_controls_1992_2024-ODACAPPED.csv')

# log transform econ_neocol_score — log1p handles zeros cleanly
df['econ_neocol_score_log'] = np.log1p(df['econ_neocol_score'] * 1e9)

# quick check
print(df['econ_neocol_score_log'].describe())
print(f"Zeros remaining: {(df['econ_neocol_score_log'] == 0).sum():,}")

df.to_csv('../data/merged/panel_with_controls_1992_2024-ODACAPPED-LOG.csv', index=False)
print('Saved.')
```
